1. Load the ratings by user information that you collected into a pandas dataframe.

```python
import pandas as pd

movie_df = pd.read_csv('movie_ratings.csv')

movie_df
```

```python
movie_df = movie_df.rename(columns = {'Unnamed: 0': 'Viewer Names'})

movie_df
```

2. Show the average ratings for each user and each movie.

```python
viewer_rating_avg = movie_df.mean(axis=1)
viewer_rating_avg
movie_df['Viewer Average Rating'] = viewer_rating_avg
movie_df
```

```python
movie_rating_avg = movie_df.mean(axis=0)
movie_rating_avg
movie_df = movie_df.append(movie_rating_avg, ignore_index = True)
movie_df['Viewer Names'].fillna('Movie Average Rating', inplace = True)
movie_df
```

3. Create a new pandas dataframe, with normalized ratings for each user. Again, show the average ratings for each user and each movie.

```python
norm_movies_df = pd.read_csv('movie_ratings.csv')
norm_movies_df = norm_movies_df.rename(columns = {'Unnamed: 0': 'Viewer Names'})

cols_to_normalize = ['Everything Everywhere All At Once','Turning Red', 'Nope', 'Hustle', 'The Unbearable Weight of Massive Talent']
norm_movies_df[cols_to_normalize] = norm_movies_df[cols_to_normalize].apply(lambda x: (x - x.min()) / (x.max() - x.min()))

norm_movies_df
```

```python
norm_viewer_rating_avg = norm_movies_df.mean(axis=1)
norm_viewer_rating_avg
norm_movies_df['Normalized Viewer Average Rating'] = norm_viewer_rating_avg
norm_movies_df
```

```python
norm_movie_rating_avg = norm_movies_df.mean(axis=0)
norm_movie_rating_avg
norm_movies_df = norm_movies_df.append(norm_movie_rating_avg, ignore_index = True)
norm_movies_df['Viewer Names'].fillna('Normalized Movie Average Rating', inplace = True)
norm_movies_df
```
